```python
from diffusers import DiffusionPipeline
import torch

# pipe = DiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16, variant="fp16", use_safetensors=True, resume_download=True)
pipe = DiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", use_auth_token=True)
pipe = pipe.to("mps")

# Recommended if your computer has < 64 GB of RAM
pipe.enable_attention_slicing()

prompt = "a photo of an astronaut riding a horse on mars"
image = pipe(prompt).images[0]
image
```
